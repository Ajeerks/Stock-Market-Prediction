```python
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
stocks=['TSLA','INTC','SOFI','IREN','NVDA']
data=yf.download(stocks,start='2023-01-01',end='2025-01-29')
close_1=data['Close']

returns=close_1.pct_change().dropna()
fig,axs=plt.subplots(3,2,figsize=(16,8),gridspec_kw={'hspace':0.3,'wspace':0.2})

axs[0,0].plot(returns['TSLA'])
axs[0,0].set_title('TSLA')
axs[0,0].set_ylim([-0.5,0.5])

axs[0,1].plot(returns['INTC'])
axs[0,1].set_title('INTC')
axs[0,1].set_ylim([-0.5,0.5])

axs[1,0].plot(returns['SOFI'])
axs[1,0].set_title('SOFI')
axs[1,0].set_ylim([-0.5,0.5])

axs[1,1].plot(returns['IREN'])
axs[1,1].set_title('IREN')
axs[1,1].set_ylim([-0.5,0.5])

axs[2,0].plot(returns['NVDA'])
axs[2,0].set_title('NVDA')
axs[2,0].set_ylim([-0.5,0.5])

fig.delaxes(axs[2,1])
plt.show()
```

```python
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
stocks=['TSLA','INTC','SOFI','IREN','NVDA']
data=yf.download(stocks,start='2023-01-01',end='2025-01-29')
close_1=data['Close']

fig,axs=plt.subplots(3,2,figsize=(16,8),gridspec_kw={'hspace':0.2,'wspace':0.1})
returns=close_1.pct_change().dropna(axis=0)

axs[0,0].hist(returns['INTC'],bins=50,range=(-0.2,0.2))
axs[0,0].set_title('INTC')

axs[0,1].hist(returns['IREN'],bins=50,range=(-0.2,0.2))
axs[0,1].set_title('IREN')

axs[1,0].hist(returns['NVDA'],bins=50,range=(-0.2,0.2))
axs[1,0].set_title('NVDA')

axs[1,1].hist(returns['SOFI'],bins=50,range=(-0.2,0.2))
axs[1,1].set_title('SOFI')

axs[2,0].hist(returns['TSLA'],bins=50,range=(-0.2,0.2))
axs[2,0].set_title('TSLA')

fig.delaxes(axs[2,1])
plt.show()
```

```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import date,timedelta
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


today=date.today()

end_date=today.strftime("%Y-%m-%d")
start_date=(today-timedelta(days=730)).strftime("%Y-%m-%d")
data=yf.download('NVDA',\
                 start=start_date,\
                    progress=False)
data.columns=data.columns.droplevel(1)     #Removing multiindex

data=data[['Close']]
data['Prediction']=data['Close'].shift(-1)
data.dropna(inplace=True)

X=data[["Close"]]
y=data[['Prediction']]

model=LinearRegression()
model.fit(X,y)

last_price=data[['Close']].iloc[-1]
print("Last Price:",last_price.values[0])

next_day=model.predict(last_price.values.reshape(1,-1))
print("Prediction Next Day price:",next_day[0])
```
